Extracting HTML code of any website using beautifulsoup

```python
import requests as r
from bs4 import BeautifulSoup as b
import pandas as pd
```

```python
url='https://www.sejda.com/pdf-editor'
page=r.get(url)
a=b(page.text,'html')
print(a)
```

```python
t=a.find('title')
print(t)
```

```python
print(t.text)
```

```python
url1='https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1'
page1=r.get(url1)
soup=b(page1.text,'html')
print(soup)
```

```python
table=soup.find('table')
print(table)
```

```python
th1=table.find_all('th')   #th data stored in list form
print(th1)
h=[i.text for i in th1]
print(h)

```

```python
td1=table.find_all('td')
print(td1)
d=[i.text for i in td1]
print(d)
d1=[d[i] for i in range(40) if i%4==0]
d2=[d[i] for i in range(40) if i%4==1]
d3=[d[i] for i in range(40) if i%4==2]
d4=[d[i] for i in range(40) if i%4==3]
```

```python
#creating datframe
df=pd.DataFrame({
    h[0]:d1,
    h[1]:d2,
    h[2]:d3,
    h[3]:d4
})
```

```python
print(df)
```

```python
#creating data frame
import pandas as pd
df1=pd.DataFrame(columns=h)
print(df1)
```

```python
rows=table.find_all('tr')
print(rows)
```

```python
ind=0
for i in rows[1:6]:
  #print(i)
  erd=i.find_all('td')
  #print(erd)
  rd=[j.text for j in erd]
  #print(rd)
  df1.loc[ind]=rd
  ind+=1

print(df1)
```
