```python
#Assignment
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df=pd.read_excel("/content/temp.xlsx")
plt.plot(df['day'], df['temperature'],marker='o')
temp=np.arange(15,df['temperature'].max()+2,5)
temp1=np.arange(1,df['day'].max()+2,2)
plt.xlabel('Days')
plt.ylabel('Temperature (C)')
plt.xticks(temp1)
plt.yticks(temp)

plt.title('Temperature analysis')
threshold = 30.0
days_above_threshold = df[df['temperature'] > threshold]['day'].count()
print(f'Number of days where temperature exceeded {threshold}C: {days_above_threshold}')
print('average temperature is',df['temperature'].mean())
print("Mininum temperature is",df['temperature'].min())
print("Maximum temperature is",df['temperature'].max())
plt.show()
```

```python
#assignment
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd



df=pd.read_excel("/content/temp.xlsx")
temp=np.arange(15,df['temperature'].max()+2,5)
temp1=np.arange(1,df['day'].max()+1,2)
c=['blue','orange','red']
labels=['blue-minumum temperature\n','orange-avg temperature\n','red-maxmum temperature\n']
colors=[]


for i in df['temperature']:
  if i<(df['temperature'].min()+3):
    colors.append(c[0%3])
  elif ((df['temperature'].min()+3)<=i) and ((df['temperature'].mean()+5)>i):
    colors.append(c[1%3])
  else:
    colors.append(c[2%3])



plt.figure(figsize=(8,4))

s=plt.scatter(df['day'], df['temperature'],color=colors)
plt.xlabel('Days')
plt.ylabel('Temperature (C)')
plt.xticks(temp1)
plt.yticks(temp)


plt.title('Temperature analysis')
threshold = 30.0
days_above_threshold = df[df['temperature'] > threshold]['day'].count()
print(f'Number of days where temperature exceeded {threshold}C: {days_above_threshold}')
print('average temperature is',df['temperature'].mean())
print("Mininum temperature is",df['temperature'].min())
print("Maximum temperature is",df['temperature'].max())
print(colors)
plt.show()
```

```python
#assignmen
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.colors import ListedColormap
from matplotlib.ticker import FixedLocator, FuncFormatter

df=pd.read_excel("/content/temp.xlsx")
colors = ['blue', 'orange', 'red']
labels = ['Minimum temperature', 'Average temperature', 'Maximum temperature']

cmap = ListedColormap(colors)
plt.figure(figsize=(8, 4))
scatter = plt.scatter(df['day'], df['temperature'], c=df['temperature'], cmap=cmap)

cbar = plt.colorbar(scatter, ticks=[df['temperature'].min(), df['temperature'].mean(), df['temperature'].max()])
cbar.ax.set_yticklabels(labels)
cbar.set_label('Temperature', rotation=270)

plt.xlabel('Days')
plt.ylabel('Temperature (C)')
plt.title('Temperature analysis')

temp = np.arange(15, df['temperature'].max() + 2, 5)
temp1 = np.arange(1, df['day'].max() + 1, 2)
plt.xticks(temp1)
plt.yticks(temp)

threshold = 30.0
days_above_threshold = (df['temperature'] > threshold).sum()
print(f'Number of days where temperature exceeded {threshold}C: {days_above_threshold}')
print('Average temperature:', df['temperature'].mean())
print('Minimum temperature:', df['temperature'].min())
print('Maximum temperature:', df['temperature'].max())
plt.show()
```
